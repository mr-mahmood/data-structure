# Understanding Queue Data Structure

## What is a Queue?
A **queue** is a linear data structure that follows the **First In, First Out (FIFO)** principle. This means that the first element added to the queue will be the first one to be removed, similar to a real-world queue, like waiting in line at a ticket counter.

## Key Characteristics of a Queue
1. **FIFO (First In, First Out):** Elements are added at one end (the rear) and removed from the other end (the front).
2. **Linear Structure:** The elements are arranged sequentially.
3. **Operations:**
   - **Enqueue:** Adding an element to the rear of the queue.
   - **Dequeue:** Removing an element from the front of the queue.

### Initial State
```
Front -> [] <- Rear
```

### Enqueue Operations
Adding elements (e.g., 10, 20, 30) to the queue:
```
Enqueue(10)
Front -> [10] <- Rear

Enqueue(20)
Front -> [10, 20] <- Rear

Enqueue(30)
Front -> [10, 20, 30] <- Rear
```

### Dequeue Operations
Removing elements from the front of the queue:
```
Dequeue()
Front -> [20, 30] <- Rear

Dequeue()
Front -> [30] <- Rear
```

## Types of Queues
1. **Simple Queue:** Operates strictly on FIFO.
2. **Circular Queue:** The last position is connected to the first, making it circular.
3. **Priority Queue:** Elements are dequeued based on priority, not just order.
4. **Deque (Double-Ended Queue):** Elements can be added or removed from both ends.

## Real-World Examples
1. **Printer Queue:** Print jobs are processed in the order they are received.
2. **Call Center Systems:** Customers are attended to based on when they joined the queue.
3. **Task Scheduling:** Tasks are executed in the order they are added.
